```python
df.drop(['employee_residence','remote_ratio','company_size'],axis=1,inplace=True)
```

```python
imputer = SimpleImputer(strategy='constant')
df2 = pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

```python
df3=df2.groupby(['work_year','job_title'])[['salary','salary_in_usd']].sum().reset_index()
```
